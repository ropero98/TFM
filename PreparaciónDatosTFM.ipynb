```python

for i in range(len(df)):
  if (df['Smoking'][i]=='Yes'):
    df['Smoking'][i]=1
  else:
    df['Smoking'][i]=0
  if (df['AlcoholDrinking'][i]=='Yes'):
    df['AlcoholDrinking'][i]=1
  else:
    df['AlcoholDrinking'][i]=0
  if (df['Stroke'][i]=='Yes'):
    df['Stroke'][i]=1
  else:
    df['Stroke'][i]=0
  if (df['DiffWalking'][i]=='Yes'):
    df['DiffWalking'][i]=1
  else:
    df['DiffWalking'][i]=0
  if (df['Sex'][i]=='Female'):
    df['Sex'][i]=1
  else:
    df['Sex'][i]=0
  if (df['AgeCategory'][i]=='18-24'):
    df['AgeCategory'][i]=0
  elif (df['AgeCategory'][i]=='25-29'):
    df['AgeCategory'][i]=1
  elif (df['AgeCategory'][i]=='30-34'):
    df['AgeCategory'][i]=2
  elif (df['AgeCategory'][i]=='35-39'):
    df['AgeCategory'][i]=3
  elif (df['AgeCategory'][i]=='40-44'):
    df['AgeCategory'][i]=4
  elif (df['AgeCategory'][i]=='45-49'):
    df['AgeCategory'][i]=5
  elif (df['AgeCategory'][i]=='50-54'):
    df['AgeCategory'][i]=6
  elif (df['AgeCategory'][i]=='55-59'):
    df['AgeCategory'][i]=7
  elif (df['AgeCategory'][i]=='60-64'):
    df['AgeCategory'][i]=8
  elif (df['AgeCategory'][i]=='65-69'):
    df['AgeCategory'][i]=9
  elif (df['AgeCategory'][i]=='70-74'):
    df['AgeCategory'][i]=10
  elif (df['AgeCategory'][i]=='75-79'):
    df['AgeCategory'][i]=11
  else:
    df['AgeCategory'][i]=12
  if (df['Race'][i]=='White'):
    df['Race'][i]=0
  elif (df['Race'][i]=='Black'):
    df['Race'][i]=1
  elif (df['Race'][i]=='Asian'):
    df['Race'][i]=2
  elif (df['Race'][i]=='American Indian/Alaskan Native'):
    df['Race'][i]=3
  elif (df['Race'][i]=='Hispanic'):
    df['Race'][i]=4
  else:
    df['Race'][i]=5
  if (df['Diabetic'][i]=='Yes'):
    df['Diabetic'][i]=0
  elif (df['Diabetic'][i]=='No'):
    df['Diabetic'][i]=1
  elif (df['Diabetic'][i]=='No, borderline diabetes'):
    df['Diabetic'][i]=2
  else:
    df['Diabetic'][i]=3
  if (df['PhysicalActivity'][i]=='Yes'):
    df['PhysicalActivity'][i]=1
  else:
    df['PhysicalActivity'][i]=0
  if (df['GenHealth'][i]=='Poor'):
    df['GenHealth'][i]=0
  elif (df['GenHealth'][i]=='Fair'):
    df['GenHealth'][i]=1
  elif (df['GenHealth'][i]=='Good'):
    df['GenHealth'][i]=2
  elif (df['GenHealth'][i]=='Very good'):
    df['GenHealth'][i]=3
  else:
    df['GenHealth'][i]=4
  if (df['Asthma'][i]=='Yes'):
    df['Asthma'][i]=1
  else:
    df['Asthma'][i]=0
  if (df['KidneyDisease'][i]=='Yes'):
    df['KidneyDisease'][i]=1
  else:
    df['KidneyDisease'][i]=0
  if (df['SkinCancer'][i]=='Yes'):
    df['SkinCancer'][i]=1
  else:
    df['SkinCancer'][i]=0
```
